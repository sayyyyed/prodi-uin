```python
import pandas as pd
import ast
import csv
```

```python
kampus = pd.read_excel('uin-dan-iain.xlsx')
```

```python
kampus.drop_duplicates(subset=['University', 'Program', 'Degree'], keep='first', inplace=True)

# Reset the index if needed
kampus.reset_index(drop=True, inplace=True)
kampus.head()
```

```python
pd.set_option('display.max_rows', None)

value_counts_series = kampus['University'].value_counts()

# Convert the Series to a DataFrame
value_counts_df = pd.DataFrame(value_counts_series)

# Rename the columns for clarity if needed
value_counts_df.columns = ['Jumlah Prodi']
value_counts_df
```

```python
value_counts_df.describe()
```

```python
kampus['norm_grade'] = kampus['norm_grade'].astype(int)
print(kampus.dtypes)
```

```python
grade_a_programs = kampus[kampus['Grade'] == 'A']

# Count the number of Grade A programs for each university
grade_a_programs_count_per_university = grade_a_programs.groupby('University').size()

value_counts_df = pd.merge(value_counts_df, grade_a_programs_count_per_university.rename('Jumlah A'), left_index=True, right_index=True, how='left')

value_counts_df['Jumlah A'].fillna(0, inplace=True)
value_counts_df['Jumlah A'] = value_counts_df['Jumlah A'].astype(int)

value_counts_df
```

```python
grade_b_programs = kampus[kampus['Grade'] == 'B']

# Count the number of Grade A programs for each university
grade_b_programs_count_per_university = grade_b_programs.groupby('University').size()

value_counts_df = pd.merge(value_counts_df, grade_b_programs_count_per_university.rename('Jumlah B'), left_index=True, right_index=True, how='left')

value_counts_df['Jumlah B'].fillna(0, inplace=True)
value_counts_df['Jumlah B'] = value_counts_df['Jumlah B'].astype(int)

value_counts_df
```

```python
grade_c_programs = kampus[kampus['Grade'] == 'C']

# Count the number of Grade A programs for each university
grade_c_programs_count_per_university = grade_c_programs.groupby('University').size()

value_counts_df = pd.merge(value_counts_df, grade_c_programs_count_per_university.rename('Jumlah C'), left_index=True, right_index=True, how='left')

value_counts_df['Jumlah C'].fillna(0, inplace=True)
value_counts_df['Jumlah C'] = value_counts_df['Jumlah C'].astype(int)

value_counts_df
```

```python
grade_D_programs = kampus[kampus['Grade'] == 'Tidak Terakreditasi']

# Count the number of Grade A programs for each university
grade_D_programs_count_per_university = grade_D_programs.groupby('University').size()

value_counts_df = pd.merge(value_counts_df, grade_D_programs_count_per_university.rename('Tidak Terakreditasi'), left_index=True, right_index=True, how='left')

value_counts_df['Tidak Terakreditasi'].fillna(0, inplace=True)
value_counts_df['Tidak Terakreditasi'] = value_counts_df['Tidak Terakreditasi'].astype(int)

value_counts_df
```

```python
value_counts_df['Poin'] = (value_counts_df['Jumlah A'] * 4) + (value_counts_df['Jumlah B'] * 3) + (value_counts_df['Jumlah C'] * 2) + value_counts_df['Tidak Terakreditasi']
value_counts_df
```

```python
value_counts_df.to_excel('poin_per_prodi.xlsx')
```
